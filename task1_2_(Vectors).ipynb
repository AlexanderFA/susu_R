```r
set.seed(123)  # задаем начальное значение генератора случайных чисел для воспроизводимости результатов
n <- 200  # задаем размер выборки
mu <- 170  # среднее значение роста
sigma <- 5  # стандартное отклонение роста
heights <- rnorm(n, mean = mu, sd = sigma)  # генерируем данные о росте
heights
```

```r
min_height <- min(heights)  # минимальное значение
max_height <- max(heights)  # максимальное значение
range_height <- diff(range(heights))  # размах

mean_height <- mean(heights)  # среднее значение

variance_biased <- var(heights)  # смещённая дисперсия
variance_unbiased <- var(heights)*(n-1)/n  # исправленная дисперсия

sd_biased <- sd(heights)  # смещённое СКО
sd_unbiased <- sd(heights) * (n-1) / n  # несмещённое СКО

cat(min_height, max_height, range_height, mean_height, variance_biased, variance_unbiased, sd_biased, sd_unbiased)
```
